```
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010]

pop = [2.519, 3.692, 5.623, 6.972]

# Add more data
year = year + [2012, 2015, 2017]
pop = pop + [7.2, 7.8, 8.3]

plt.plot(year, pop)

# Add x label
plt.xlabel('Year')
# Add y label
plt.ylabel('Population')
# Add title
plt.title('World population')
# Modify the y axis value, each number will bind with a name
plt.yticks([0, 2, 4, 6, 8, 10], ['0', '2B', '4B', '6B', '8B', '10B'])

# this render the plot
plt.show()
```
